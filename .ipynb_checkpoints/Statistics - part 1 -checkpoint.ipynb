## Data distributions
The distribution of a statistical data set (or a population) is a listing or function showing all the possible values (or intervals) of the data and how often they occur. When a distribution of categorical data is organized, you see the number or percentage of individuals in each group.


## What define shape of distributions
- Spread out (e.g. dispersion, variability, scatter),
- Where the mean lies,
- What the range of the data set is,

#### Number of peaks
The peaks are usually called modes; Sometimes they are called local maximums, but that term is more commonly used in calculus. The mode tells you that the data count is higher in these areas than in any other areas on the graph.
 - **Unimodal **A unimodal distribution has one mode. A single peak can take on many shapes (e.g. very tall and thin or very squat and fat). One of the most common types of unimodal distributions is the normal distribution, sometimes called the bell curve because its shape looks like a bell.



 - **bimodal** A bimodal distribution has two modes.
 

 - **multimodel** A multimodal distribution has three or more modes.


- **Uniform distribution** If a data set has no clear peaks (i.e. the whole graph looks flat), its called a uniform distribution.

### Summarizing quantitative data

### Quartile
A quartile is a type of quantile. The first quartile (Q1) is defined as the middle number between the smallest number and the median of the data set. ... The third quartile (Q3) is the middle value between the median and the highest value of the data set.


```python
def find_quartile(sortedarray):
    getlength = len(sortedarray)
    if getlength%2==0:
        #means evens number length
        getindex2 = (int((getlength+1)/2)-1)+1
        getindex1 = (int((getlength+1)/2)-1)
        q1  =int((len(sortedarray[0:getindex1+1])+1)/2)-1
        q3  =getindex2+int((len(sortedarray[getindex2:])+1)/2)-1
        return sortedarray[q1],sortedarray[q3]
    else:
        #means odd number length
        getindex = (int((getlength+1)/2)-1)
        q2 = getindex
        q1 = int(len(sortedarray[0:getindex-1])/2)
        q3 = getindex+int(len(sortedarray[getindex:])/2)
        return sortedarray[q1],sortedarray[q3]
```

```python
find_quartile([1,2,3,4,5,6,7])
```

```python
find_quartile([1,3,3,4,5,6,6,7,8,8])
```

### measures of spread
Measures of spread describe how similar or varied the set of observed values are for a particular variable (data item). Measures of spread include the range, quartiles and the interquartile range, variance and standard deviation

## Variances of the population
 The formula of population variance is sigma squared equals the sum of x minus the mean squared divided by N
 ## Variances of the sample
 The formula of sample variance is sigma squared equals the sum of x minus the mean squared divided by n


```python
# calculating the variance populations
populationmean =0
for i in datapoints:
    populationmean = populationmean+i
populationvariance =0
for i in datapoints:
    populationvariance =populationvariance+ (i-populationmean)**2
populationvariance/len(datapoints)
```

```python
# calculating the variance samples
makingsample =[datapoints[9],datapoints[13],datapoints[3],datapoints[7],datapoints[12],datapoints[5]] #making 6samples randomly
samplemean =0
for i in makingsample:
    samplemean = samplemean+i
samplevariance =0
for i in makingsample:
    samplevariance =samplevariance+ (i-samplemean)**2
samplevariance/(len(makingsample)-1)
```

#### sample variance devide by n-1 for the unbiased estimate of variances
#### why use n-1 in sample variances ??
Bessel's correction is the use of n  1 instead of n in the formula for the sample variance and sample standard deviation, where n is the number of observations in a sample. This method corrects the bias in the estimation of the population variance. ... gives an unbiased estimator of the population variance.
for more goes to this link

### Rule of 1.5 X IQR (where IQR is interquartile range)
Outliers here are defined as observations that fall below Q1  1.5 IQR or above Q3 + 1.5 IQR. In a boxplot, the highest and lowest occurring value within this limit are indicated by whiskers of the box (frequently with an additional bar at the end of the whisker) and any outliers as individual points.

```python
#finding the outliers with 1.5 X IQR
q1,q3 =find_quartile(datapoints)
for i in datapoints:
    if i < (q1 - 1.5*(q3-q1)) or i > (q3 + 1.5*(q3-q1)):
        print(str(i) + "is the outliers for this data")
```

### Mean absolute deviations
The average absolute deviation (or mean absolute deviation) of a data set is the average of the absolute deviations from a central point. It is a summary statistic of statistical dispersion or variability. In this general form, the central point can be the mean, median, mode, or the result of another measure of central tendency.


```python
# finding the mean absolute deviations
populationmean =0
for i in datapoints:
    populationmean = populationmean+i
populationvariance =0
for i in datapoints:
    populationvariance =populationvariance+ np.abs(i-populationmean)
populationvariance/len(datapoints)
```

### Percentiles
A percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations fall. For example, the 20th percentile is the value below which 20% of the observations may be found.
<br>
Percentile is in everyday use, but there is no universal definition for it. The most common definition of a percentile is a number where a certain percentage of scores fall below that number. You might know that you scored 67 out of 90 on a test. But that figure has no real meaning unless you know what percentile you fall into. If you know that your score is in the 90th percentile, that means you scored better than 90% of people who took the test.

### Percentile rank
The percentile rank of a score is the percentage of scores in its frequency distribution that are equal to or lower than it.
- The nth percentile is the lowest score that is greater than a certain percentage (n) of the scores. In this example, our n is 25, so were looking for the lowest score that is greater than 25%.
- The nth percentile is the smallest score that is greater than or equal to a certain percentage of the scores. To rephrase this, its the percentage of data that falls at or below a certain observation. This is the definition used in AP statistics. In this example, the 25th percentile is the score thats greater or equal to 25% of the scores.

```python
scoresvsrank =[30,33,43,53,56,67,68,72]
rankvsscores =[1,2,3,4,5,6,7,8]
```

```python
## how to find the percentiles
## sample questions -  Find out where the 25th percentile is in the above list.
# The lowest score that is greater than 25% of the scores
# first calculate the the commulative frequency
scorescommulativefrequency =[]
sumt = 0
for i in scoresvsrank:
    sumt =sumt+i
    scorescommulativefrequency.append(sumt)
```

```python
scorescommulativefrequency
```

```python
# commulative relative frequency gives the percentages of data at the points
scoresrelativefrequency=[]
sumt = 0
for i in scoresvsrank:
    sumt = sumt + (i/scorescommulativefrequency[-1])
    scoresrelativefrequency.append(sumt)
```

```python
scoresrelativefrequency
```

```python
print(len(scoresrelativefrequency))
for i in range(len(scoresrelativefrequency)):
    if scoresrelativefrequency[i]>(25/100):
        print(rankvsscores[i])
```

### Percentile Range

- Calculate the 90th percentile using the above steps.

### what is a correlation?
- When the yyy variable tends to increase as the xxx variable increases, we say there is a positive correlation between the variables.
- When the yyy variable tends to decrease as the xxx variable increases, we say there is a negative correlation between the variables.
- When there is no clear relationship between the two variables, we say there is no correlation between the two variables.


### What is a correlation coefficient?
The correlation coefficient rrr measures the direction and strength of a linear relationship. Calculating rrr is pretty complex, so we usually rely on technology for the computations. We focus on understanding what rrr says about a scatterplot.
Here are some facts about rrr:
- It always has a value between -1 and 1.
- Strong positive linear relationships have values of r closer to 1.
- Strong negative linear relationships have values of r closer to -1.
- Weaker relationships have values of r closer to 0.


```python
z_scores_x = []
z_scores_y =[]
xmean = np.mean(x)
ymean =np.mean(y)
xstd =np.std(x)
ystd =np.std(y)
for i in range(len(list(x))):
    z_scores_x.append((x[i]-xmean)/xstd)
    z_scores_y.append((y[i]-ymean)/ystd)
```

```python
sumt =0
for i in range(len(z_scores_x)):
    sumt = sumt+(z_scores_x[i]*z_scores_y[i])
```

### Covariance between two random variables . 
In probability theory and statistics, covariance is a measure of the joint variability of two random variables.<br>
Covariance is a measure of how much two random variables vary together. Its similar to variance, but where variance tells you how a single variable varies, co variance tells you how two variables vary together.<br>

<center>COV(X,Y) = E[(X-E(X))(Y-E(Y)]</center>
<center>
The formula is:<br>
Cov(X,Y) =  E((X-)E(Y-)) / n-1 where:<br>
X is a random variable<br>
E(X) =  is the expected value (the mean) of the random variable X and<br>
E(Y) =  is the expected value (the mean) of the random variable Y<br>
n = the number of items in the data set<br>
    </center>

```python
x_values =[2.1, 2.5, 3.6, 4.0 ]
y_values =[8, 10, 12, 14]
```

```python
# finding covariances
xmean = np.mean(x_values)
ymean =np.mean(y_values)
sumt =0
for i in range(len(x_values)):
    sumt =sumt+((x_values[i]-xmean)*(y_values[i]-ymean))
```

```python
sumt/(len(x_values)-1) #The result is positive, meaning that the variables are positively related.
```

A large covariance can mean a strong relationship between variables. However, you cant compare variances over data sets with different scales (like pounds and inches). A weak covariance in one data set may be a strong one in a different data set with different scales.
<br>
***
The main problem with interpretation is that the wide range of results that it takes on makes it hard to interpret. For example, your data set could return a value of 3, or 3,000. This wide range of values is cause by a simple fact; The larger the X and Y values, the larger the covariance. A value of 300 tells us that the variables are correlated, but unlike the correlation coefficient, that number doesnt tell us exactly how strong that relationship is. The problem can be fixed by dividing the covariance by the standard deviation to get the correlation coefficient.
Corr(X,Y) = Cov(X,Y) / XY
